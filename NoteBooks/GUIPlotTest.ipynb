```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
 
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'h', area_thresh = 1000.0,
    llcrnrlon=-122.6445, llcrnrlat=37.1897,
    urcrnrlon=-121.5871, urcrnrlat=38.2033)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='yellow')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

lons = [-121.8817596, -122.266083, -121.8576448]
lats = [37.40842, 37.8123695, 37.3438417]
x,y = my_map(lons, lats)
my_map.plot(x, y, 'bo', markersize=5)
 
labels = ['Xian Kitchen, 0.63', 'dosa By DOSA, 0.79', 'Burger King, 0.32']
for label, xpt, ypt in zip(labels, x, y):
    plt.text(xpt, ypt, label)

plt.show()
```
