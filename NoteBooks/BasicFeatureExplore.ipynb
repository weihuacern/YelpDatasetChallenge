```python
# first play with world cloud, bonus...
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud

df_review_sub = df_review.iloc[0:1000, :]
cloud = WordCloud(width=1500, height=1500).generate(" ".join(df_review_sub['text'].astype(str)))
plt.figure(figsize=(20, 20))
plt.imshow(cloud)
plt.axis('off')
```

```python
import seaborn as sns
# plot price vs item condition
#pricecut_lo = 100
#pricecut_hi = 500
#df_train_tmp = df_train[ (df_train['price'] > pricecut_lo) & (df_train['price'] < pricecut_hi) ]
my_plot = []

#print(df_review['stars'].nunique())
for i in df_review['stars'].unique():
    my_plot.append(df_review[df_review['stars']==i]['text'].str.len())

#print(len(my_plot))
fig, axes = plt.subplots(figsize=(20, 20))
bp = axes.boxplot(my_plot, vert=True, patch_artist=True, labels=range(1,6)) 

colors = ['cyan', 'lightblue', 'lightgreen', 'tan', 'pink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

axes.yaxis.grid(True)

plt.title('BoxPlot text length X stars', fontsize=20)
plt.xlabel('Stars', fontsize=20)
plt.ylabel('Text Length', fontsize=20)
plt.show()
```
